```python
import os
import shutil
import pandas as pd

# 업로드된 CSV 파일 읽기
file_path = 'Other_model.csv'
data_df = pd.read_csv(file_path,encoding='cp949')

# CSV 파일에서 'pid' 컬럼 읽기
pids = data_df['id'].tolist()
```

```python
# 이미지 폴더의 기본 경로
base_image_path = "../Image"

# 복사할 위치 지정
destination_base_path = "./Image_folder"

# pids 리스트의 각 pid에 대해 폴더 복제
for target_pid in pids:
    # source_folder_path를 /image/{pid}로 설정
    source_folder_path = os.path.join(base_image_path, target_pid)
    
    if os.path.exists(source_folder_path):
        # 목적지 폴더 경로 설정
        destination_folder_path = os.path.join(destination_base_path, os.path.basename(source_folder_path))
        
        # 폴더 복제
        shutil.copytree(source_folder_path, destination_folder_path)
        print(f"폴더가 성공적으로 복제되었습니다: {destination_folder_path}")
    else:
        print(f"해당 pid({target_pid})에 일치하는 폴더를 찾을 수 없습니다.")
```
